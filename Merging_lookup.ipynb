```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_holds = pd.read_csv("lansa.csv")
df_engine = pd.read_csv("engine list.csv")

df_holds.head()
```

```python
df_engine.rename(columns = {'Suspected engine list':'ENGINE #'}, inplace = True)
```

```python
df_engine.head()
```

```python
df_total = pd.merge(df_engine, df_holds, how = "inner", on = "ENGINE #")
```

```python
df_total
```

```python
df_total.to_csv("engines_conc.csv")
```
